```python
import pandas as pd
base = pd.read_csv('trees.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
#Histograma
base.hist(column='Height', bins='auto')
plt.title('√Årvores')
plt.xlabel('Altura')
plt.ylabel('Frequencia')
plt.show()
```

```python
#Densidade
sns.distplot(base['Height'], hist=True, kde= True, bins='auto', color='green', hist_kws={'edgecolor':'green'})
plt.title('√Årvores')
plt.xlabel('Altura')
plt.ylabel('Frequencia')
plt.show()
plt.show()
```

```python
#Histograma e Densidade - com Seaborn
sns.distplot(base.Volume, bins='auto', axlabel='Volume').set_title('Arvores')
plt.show()
```

```python
#Dispers√£o
plt.scatter(base.Girth, base.Volume, color='seagreen', facecolors='none', marker='.')
plt.title('√Årvores')
plt.xlabel('Frequencia')
plt.ylabel('Volume')
plt.show()
```

```python
#Dispers√£o sem sobreposi√ß√£o
sns.regplot(base.Girth, base.Volume, color='seagreen', data=base, x_jitter=1.5, fit_reg=False)
plt.title('√Årvores')
plt.xlabel('Frequencia')
plt.ylabel('Volume')
plt.show()
```

```python
#Dispers√£o com Legendas
base= pd.read_csv('co2.csv')
)
```

```python
x= base.conc
y= base.uptake
unicos= list(set(base.Treatment))

for i in range (len(unicos)):
    indice= base.Treatment==unicos[i]
    plt.scatter(x[indice], y[indice], label=unicos[i])
    plt.legend(loc='lower right')
    plt.show()
    
```

```python
#Dividindo as telas - subgraficos

plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(base.Girth, base.Volume)
plt.title('√Årvores', loc='center')
plt.xlabel('Frequencia')
plt.ylabel('Volume')


plt.subplot(2,2,2)
plt.scatter(base.Girth, base.Height)
plt.title('√Årvores', loc='center')
plt.xlabel('Frequencia')
plt.ylabel('Altura')
plt.show()
```

```python
#Boxplot
plt.boxplot(base.Volume, vert=False, showfliers=True, notch=True, patch_artist=True)
plt.title('√Årvores')
plt.xlabel('Volume')
plt.show()
```

```python
base.boxplot(column=['Volume', 'Girth'], showfliers=True, patch_artist=True)
plt.title('√Årvores')
plt.show()
```

```python
#Boxplot - usando seaborn
sns.boxplot(data=base).set_title('Arvores')
plt.show()
```
