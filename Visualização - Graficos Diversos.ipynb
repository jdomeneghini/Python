```python
import pandas as pd
base = pd.read_csv('trees.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
#Histograma
base.hist(column='Height', bins='auto')
plt.title('Árvores')
plt.xlabel('Altura')
plt.ylabel('Frequencia')
plt.show()
```

```python
#Densidade
sns.distplot(base['Height'], hist=True, kde= True, bins='auto', color='green', hist_kws={'edgecolor':'green'})
plt.title('Árvores')
plt.xlabel('Altura')
plt.ylabel('Frequencia')
plt.show()
plt.show()
```

```python
#Histograma e Densidade - com Seaborn
sns.distplot(base.Volume, bins='auto', axlabel='Volume').set_title('Arvores')
plt.show()
```

```python
#Dispersão
plt.scatter(base.Girth, base.Volume, color='seagreen', facecolors='none', marker='.')
plt.title('Árvores')
plt.xlabel('Frequencia')
plt.ylabel('Volume')
plt.show()
```

```python
#Dispersão sem sobreposição
sns.regplot(base.Girth, base.Volume, color='seagreen', data=base, x_jitter=1.5, fit_reg=False)
plt.title('Árvores')
plt.xlabel('Frequencia')
plt.ylabel('Volume')
plt.show()
```

```python
#Dispersão com Legendas
base= pd.read_csv('co2.csv')
```

```python
x= base.conc
y= base.uptake
unicos= list(set(base.Treatment))

for i in range (len(unicos)):
    indice= base.Treatment==unicos[i]
    plt.scatter(x[indice], y[indice], label=unicos[i])
    plt.legend(loc='lower right')
    plt.show()
    
```

```python
#Dividindo as telas - subgraficos

plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(base.Girth, base.Volume)
plt.title('Árvores', loc='center')
plt.xlabel('Frequencia')
plt.ylabel('Volume')


plt.subplot(2,2,2)
plt.scatter(base.Girth, base.Height)
plt.title('Árvores', loc='center')
plt.xlabel('Frequencia')
plt.ylabel('Altura')
plt.tight_layout()
```

```python
#Boxplot
plt.boxplot(base.Volume, vert=False, showfliers=True, notch=True, patch_artist=True)
plt.title('Árvores')
plt.xlabel('Volume')
plt.show()
```

```python
base.boxplot(column=['Volume', 'Girth'], showfliers=True, patch_artist=True)
plt.title('Árvores')
plt.show()
```

```python
#Boxplot - usando seaborn
sns.boxplot(data=base).set_title('Arvores')
plt.show()
```

```python
sns.boxplot(base.Volume).set_title('Arvores')
plt.show()
```

```python
#Dispersao - Scatter Seaborn
base= pd.read_csv('co2.csv')
```

```python
base.head()
```

```python
base['Type'].value_counts()
base['Treatment'].value_counts()
```

```python
sns.scatterplot(base.conc, base.uptake, hue=base.Type)
plt.title('')
```

```python
q=base.loc[base['Type']=='Quebec']
m= base.loc[base['Type']=='Mississippi']
```

```python
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(q.conc, q.uptake).set_title('Quebec')

plt.subplot(1,2,2)
sns.scatterplot(m.conc, m.uptake).set_title('Mississippi')
plt.tight_layout()
```

```python
ch=base.loc[base['Treatment']=='chilled']
nch= base.loc[base['Treatment']== 'nonchilled']
```

```python
plt.figure()
plt.subplot(1,2,1)
sns.scatterplot(ch.conc, ch.uptake).set_title('Chilled')

plt.subplot(1,2,2)
sns.scatterplot(nch.conc, nch.uptake).set_title('NonChilled')
plt.tight_layout()


```
