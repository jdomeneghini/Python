```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use ('ggplot')
```

```python
#Leitura do arquivo
videogames= pd.read_csv('vgsales.csv')
```

```python
#Resumo de informações em todas as colunas
videogames.describe()
```

```python
#Tipo de dado em cada coluna
videogames.dtypes
```

```python
#Quantidade de linhas e colunas no Dataframe
videogames.shape
```

```python
#Renomeando as colunas
videogames.columns= ['Ranking', 'Nome', 'Plataforma', 'Ano', 'Gênero', 
                     'Editora', 'Vendas América do Norte', 'Vendas EUA', 
                     'Vendas Japão','Outras Vendas', 'Vendas Global']
```

```python
#Exibindo as 10 primeiras linhas do Dataframe
videogames.head(10)
```

```python
#Verificando linhas onde não há ano de lançamento definido
videogames[videogames['Ano'].isnull()].head()
```

```python
#Contagem de jogos lançados por cada plataforma
videogames['Plataforma'].value_counts()
```

```python
#Mostrando em gráfico
titulos_lancados= videogames['Plataforma'].value_counts()
titulos_lancados.plot()
```

```python
#Criando um grafico com somente uma linha de código
videogames['Plataforma'].value_counts().head(10).plot(kind='bar', 
                                                     figsize=(11,5), 
                                                     grid=False,
                                                    rot=0, color='green')

#Enfeitando o grafico
plt.title('Os 10 videogames com mais títulos lançados')
plt.xlabel('Videogame')
plt.ylabel('Quantidades de jogos lançados')
plt.show()
```

```python
#Os 10 jogos mais vendidos da história
top_10_vendas= videogames[['Nome', 'Vendas Global']].head(10).set_index('Nome').sort_values('Vendas Global', ascending=True)
top_10_vendas.plot(kind='barh', figsize=(11,7), grid=False, color='darkred', legend=False)

plt.title('Os 10 jogos mais vendidos da história')
plt.xlabel('Total de Vendas (em milhões de doláres)')
plt.show()
```

```python
#Criando um mapa de calor
#Cruzando duas variaveis com função .crosstab()

crosstab_vg= pd.crosstab(videogames['Plataforma'],videogames['Gênero'])
crosstab_vg.head()                        
```

```python
#Somando os jogos lançados para cada console
crosstab_vg['Total']= crosstab_vg.sum(axis=1)
crosstab_vg.head()
```

```python
#Montando o gráfico
top10_plataformas= crosstab_vg[crosstab_vg['Total']>1000].sort_values('Total',ascending=False)
top10_final= top10_plataformas.append(pd.DataFrame(top10_plataformas.sum(), columns=['Total']).T, ignore_index=False)

sns.set(font_scale=1)
plt.figure(figsize=(18,9))
sns.heatmap(top10_final, annot=True, vmax=top10_final.loc[:'PS', :'Strategy'].values.max(), 
            vmin=top10_final.loc[:, :'Strategy'].values.min(), fmt='d')
plt.title('Quantidade de Títulos por Gênero e Console')
plt.xlabel('Gênero')
plt.ylabel('Console')
plt.show()
```
